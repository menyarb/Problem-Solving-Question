```python
import re

def somme_valeurs_etalonnage(fichier):
    somme = 0
    valeurs = []  
   
    with open(fichier, 'r') as file:
        for ligne in file:
          
            ligne = ligne.strip()

            
            chiffres = re.findall(r'\d', ligne)

            if len(chiffres) >= 2:
              
                valeur = int(chiffres[0] + chiffres[-1])
                valeurs.append(valeur) 
                somme += valeur

    return somme, valeurs


fichier = "document.txt"


somme_totale, valeurs_etalonnage = somme_valeurs_etalonnage(fichier)


valeurs_str = ", ".join(map(str, valeurs_etalonnage)) 
print(f"Ces lignes sont {valeurs_str}.")
print(f"La somme de toutes les valeurs d'étalonnage est : {somme_totale}")
```
